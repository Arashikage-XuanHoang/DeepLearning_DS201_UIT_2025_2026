# RNN đơn giản (one-to-one, many-to-one)

```python
import tensorflow as tf

# Cấu hình mô hình RNN nhiều lớp (many-to-one)
num_classes = 5  # ví dụ 5 lớp phân loại
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=32),   # Embedding từ từ thành vector
    tf.keras.layers.SimpleRNN(64),                              # RNN layer với 64 đơn vị ẩn
    tf.keras.layers.Dense(num_classes, activation='softmax')    # Lớp output với softmax
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

# LSTM

```python
import tensorflow as tf

# Ví dụ LSTM cho phân loại chuỗi (many-to-one)
num_classes = 3  # ví dụ 3 lớp phân loại
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=64),   # Embedding với 64 chiều
    tf.keras.layers.LSTM(128),                                  # LSTM layer với 128 đơn vị ẩn
    tf.keras.layers.Dense(num_classes, activation='softmax')    # Output softmax
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

# GRU

```python
import tensorflow as tf

# Ví dụ GRU cho phân loại chuỗi
num_classes = 4
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=64),   # Embedding
    tf.keras.layers.GRU(64),                                    # GRU layer với 64 đơn vị ẩn
    tf.keras.layers.Dense(num_classes, activation='softmax')    # Output softmax
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

# BiLSTM

```python
import tensorflow as tf

# Ví dụ BiLSTM cho phân loại chuỗi (many-to-one)
num_classes = 2
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=64),     # Embedding
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),      # BiLSTM với 64 đơn vị (32 mỗi chiều)
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

# Mạng LSTM xếp chồng

```python
import tensorflow as tf

# Ví dụ Stacked LSTM (2 tầng)
num_classes = 5
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=64),
    tf.keras.layers.LSTM(128, return_sequences=True),  # LSTM tầng 1 (xuất ra chuỗi)
    tf.keras.layers.LSTM(128),                         # LSTM tầng 2 (xuất ra ẩn cuối)
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
import os
import pandas as pd
from collections import Counter

# Đường dẫn file
data_dir = "/content/UIT-VSFC"

def load_split(split_name):
  sent_path = os.path.join(data_dir, split_name, "sents.txt")
  senti_path = os.path.join(data_dir, split_name, "sentiments.txt")
  topic_path = os.path.join(data_dir, split_name, "topics.txt")

  # Đọc file
  with open(sent_path, 'r', encoding='utf-8') as f:
      sentences = [line.strip() for line in f if line.strip()]
  with open(senti_path, 'r', encoding='utf-8') as f:
      sentiments = [int(line.strip()) for line in f if line.strip()]
  with open(topic_path, 'r', encoding='utf-8') as f:
      topics = [line.strip() for line in f if line.strip()]

  df = pd.DataFrame({
      'sentence': sentences,
      'sentiment': sentiments,
      'topic': topics
  })
  # Map nhãn sentiment sang chữ
  df['sentiment_label'] = df['sentiment'].map({2: 'Positive', 1: 'Neutral', 0: 'Negative'})
  return df
```

```python
import re
def count_words(sentence):
    # Loại bỏ toàn bộ dấu câu trước khi đếm
    clean = re.sub(r'[^\w\s]', ' ', str(sentence))   # thay . , ! ? ... bằng khoảng trắng
    clean = re.sub(r'\s+', ' ', clean).strip()       # xóa khoảng trắng thừa
    return len(clean.split())                        # đếm số từ
```

```python
import re
import unicodedata
from underthesea import word_tokenize


# Load stopwords (bộ stopwords tiếng Việt cơ bản)

stopwords = set([
    "và", "là", "có", "cho", "rất", "nhiều", "được", "bị", "của", "không",
    "nhưng", "với", "thì", "lại", "một", "các", "như", "để", "khi", "đã",
    "ở", "trong", "ra", "vào", "này", "kia"
])


# Hàm preprocessing

def preprocess_text(text):
    # --- Chuẩn hóa unicode ---
    text = unicodedata.normalize("NFC", text)

    # --- Lowercase ---
    text = text.lower()

    # --- Xóa emoji ---
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags
        "]+",
        flags=re.UNICODE
    )
    text = emoji_pattern.sub("", text)

    # --- Loại bỏ ký tự đặc biệt ---
    text = re.sub(r"[^a-záàảãạăắằẳẵặâấầẩẫậđéèẻẽẹêếềểễệ"
                  r"íìỉĩịóòỏõọôốồổỗộơớờởỡợ"
                  r"úùủũụưứừửữựýỳỷỹỵ\s]", " ", text)

    # --- Tokenization tiếng Việt ---
    text = word_tokenize(text, format="text")  # nối từ ghép bằng dấu gạch dưới

    # --- Bỏ stopwords ---
    tokens = text.split()
    tokens = [t for t in tokens if t not in stopwords]

    # --- Chuẩn hóa số ---
    tokens = ["<number>" if t.isdigit() else t for t in tokens]

    # --- Xóa khoảng trắng thừa ---
    text = " ".join(tokens).strip()

    return text


# Ví dụ chạy thử

raw = "Thầy GIẢNG VIÊN dạy RẤT nhiệt tình 😊, bài giảng dễ hiểu nhưng hơi nhanh!!!"

processed = preprocess_text(raw)

print("Before:\n", raw)
print("\nAfter:\n", processed)
```

```python
# Tính số token trong từng câu

df['token_count'] = df['tokens'].apply(len)

# Min/max/mean trong tập df

print("Min token count:", df['token_count'].min())
print("Max token count:", df['token_count'].max())
print("Mean token count:", df['token_count'].mean())
```

```python
import numpy as np
# Padding tối ưu

all_lens = df["token_count"]
print(f"Tổng số câu       : {len(all_lens):,}")
print(f"Độ dài trung bình : {all_lens.mean():.2f} token")
print(f"Độ dài trung vị   : {np.median(all_lens):.0f} token")
print(f"Độ dài tối đa     : {all_lens.max()} token")
print(f"95% câu ≤         : {int(np.quantile(all_lens, 0.95))} token")
print(f"99% câu ≤         : {int(np.quantile(all_lens, 0.99))} token")
print(f"99.9% câu ≤       : {int(np.quantile(all_lens, 0.999))} token")
```

```python
# Tổng số sample có độ dài > 69 token
print(f"Tổng số sample có độ dài > 69 token: {len(df[df['token_count'] > 69])}")
```

- Chọn max_length = 69 làm giá trị padding tối ưu trong toàn bộ các thí nghiệm.
Với max_length = 69, có thể giữ nguyên 99% dữ liệu gốc (chỉ cắt 1% số câu rất dài)

- Có 98 câu dài hơn padding tối ưu gây khó khăn cho LSTM.
